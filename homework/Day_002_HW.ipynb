```python
'檢查行名稱及其相應的數據類型'
app_train.info()
```

```python
'Search NaN，用來處理資料NaN'
app_train_drop = app_train.drop(columns=['NAME_CONTRACT_STATUS'])
np.isnan(app_train_drop)               #去除object type 才可以用.isnan()找 NaN
```

```python
'總數，平均值，標準差，最小值和最大值以及數據的分位數，用來看資料離散度'
app_train_drop.describe()
```

```python
'使用matplotlib中的boxplot可視化平均值，最小值和最大值以及數據的分位數，用來看資料離散度'
import matplotlib.pyplot as plt
plt.boxplot(app_train_drop.AMT_TOTAL_RECEIVABLE, sym ="o", whis = 1.5)
```

```python
'了解CNT_DRAWINGS_CURRENT數置分布'
app_train.CNT_DRAWINGS_CURRENT.value_counts()
```

```python
'使用panda.corr（）函數找到相關性，並使用seaborn中的熱圖可視化相關矩陣'
import seaborn as sns
sns.set()
ax = sns.heatmap(app_train.corr(), vmin=0, vmax=1, center=0)          #, annot=True
```
