將下列部分資料片段 sub_train 使用 One Hot encoding, 並觀察轉換前後的欄位數量 (使用 shape) 與欄位名稱 (使用 head) 變化

```python
sub_train = pd.DataFrame(app_train['WEEKDAY_APPR_PROCESS_START'])
sub_train.head()
```

```python
# sub_train.as_matrix()
# sub_train.values
asub_train = np.array(sub_train)
asub_train[:,0]
```

```python
"""
Your Code Here
"""
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
LabelEn = LabelEncoder().fit_transform(asub_train[:,0])
print(LabelEn.shape)
print(LabelEn)
```

```python
OneHotEn = OneHotEncoder().fit_transform(LabelEn.reshape(-1,1))
print(OneHotEn.shape)
print(type(OneHotEn))
print(OneHotEn.toarray())
```

```python
sub_train=pd.DataFrame(OneHotEn.toarray(), columns=['FRIDAY','MONDAY','SATURDAY','SUNDAY','THURSDAY','TUESDAY','WEDNESDAY'])
sub_train.head()
```
