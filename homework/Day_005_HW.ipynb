```python
import matplotlib.pyplot as plt
plt.boxplot(app_train.AMT_CREDIT, sym ="o", whis = 1.5)
```

```python
#觀察平均值，最小值和最大值以及數據的分位數，用來看資料離散度 
fig, axs = plt.subplots(3)

axs[0].boxplot(app_train.CNT_CHILDREN)
axs[0].set_title('CNT_CHILDREN')

axs[1].boxplot(app_train.AMT_INCOME_TOTAL, 1)
axs[1].set_title('AMT_INCOME_TOTAL')

axs[2].boxplot(app_train.AMT_CREDIT, 0, 'gD')
axs[2].set_title('AMT_CREDIT')

fig.subplots_adjust(left=0.08, right=0.58, bottom=0.05, top=1.9,
                    hspace=0.4, wspace=0.3)
```

```python
fig, axs = plt.subplots(2, sharey=True, tight_layout=True)

axs[0].set_ylim(0, 7500)
axs[0].hist(app_train.DAYS_BIRTH, bins=100)
axs[0].set_title('DAYS_BIRTH') 
axs[1].set_ylim(0, 30000)
axs[1].hist(app_train.AMT_CREDIT, bins=100)
axs[1].set_title('AMT_CREDIT') 
```
